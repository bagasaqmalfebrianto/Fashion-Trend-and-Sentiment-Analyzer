```python
import random
import pandas as pd
from datetime import datetime, timedelta

# Helper Functions
def random_date(start, end):
    return start + timedelta(seconds=random.randint(0, int((end - start).total_seconds())))

def random_float(min_value, max_value):
    return random.uniform(min_value, max_value)

# Define start and end date for the random date generation
start_date = datetime(2020, 1, 1)
end_date = datetime(2024, 1, 1)

total_records = 1000000  # Example record count

#  Mapping of categories to subcategories, materials, keywords, and hashtags
category_subcategories_materials_keywords_hashtags = {
    "Footwear": {
        "subcategories": ["Running Shoes", "Sneakers", "Boots", "Sandals", "Flip Flops", "Formal Shoes", "Hiking Boots", "Slippers", "Loafers", "High Heels", "Wedges", "Ballet Flats", "Espadrilles"],
        "materials": ["Mesh", "Leather", "Rubber", "Suede", "Canvas", "Synthetic", "Neoprene", "Nylon"],
        "keywords": ["Comfortable", "Durable", "Lightweight", "Breathable", "Trendy", "Fashionable", "Cushioned", "Outdoor", "Running", "Waterproof", "Stylish", "Slip-on", "Sneakers", "High Heels", "Elegant"],
        "hashtags": ["#fashion", "#footwear", "#trendy", "#comfortable", "#sportswear", "#running", "#waterproof", "#stylish", "#outdoor", "#sneakers"]
    },
    "Apparel": {
        "subcategories": ["T-Shirts", "Hoodies", "Sweatshirts", "Jackets", "Blazers", "Pants", "Jeans", "Shorts", "Skirts", "Dresses", "Suits", "Tops", "Sweaters", "Activewear", "Lingerie", "Maternity Wear", "Pajamas", "Tunics"],
        "materials": ["Cotton", "Polyester", "Wool", "Silk", "Linen", "Denim", "Leather", "Spandex", "Velvet"],
        "keywords": ["Stylish", "Comfortable", "Trendy", "Casual", "Formal", "Cozy", "Athletic", "Warm", "Soft", "Fashion", "Luxury"],
        "hashtags": ["#fashion", "#apparel", "#stylish", "#comfortable", "#cozy", "#luxury", "#athletic", "#casual", "#denim", "#tshirts"]
    },
    "Accessories": {
        "subcategories": ["Watches", "Hats", "Bags", "Scarves", "Gloves", "Belts", "Sunglasses", "Wallets", "Jewelry", "Necklaces", "Earrings", "Bracelets", "Rings", "Hair Accessories", "Headbands", "Ties"],
        "materials": ["Leather", "Cotton", "Stainless Steel", "Plastic", "Rubber", "Acrylic", "Fabric", "Wood"],
        "keywords": ["Chic", "Stylish", "Functional", "Durable", "Fashionable", "Practical", "Elegant", "Trendy", "Luxury", "Comfortable", "Designer"],
        "hashtags": ["#accessories", "#fashion", "#luxury", "#chic", "#stylish", "#bags", "#elegant", "#designer", "#trendy", "#watches"]
    },
    "Outerwear": {
        "subcategories": ["Coats", "Trench Coats", "Parkas", "Raincoats", "Cardigans", "Ponchos", "Bomber Jackets", "Peacoats", "Blousons"],
        "materials": ["Wool", "Leather", "Polyester", "Nylon", "Fleece", "Cotton", "Linen"],
        "keywords": ["Warm", "Stylish", "Durable", "Comfortable", "Waterproof", "Trendsetter", "Classic", "Modern", "Versatile", "Fashion-forward"],
        "hashtags": ["#outerwear", "#coats", "#fashion", "#stylish", "#warm", "#trendsetter", "#bomberjacket", "#raincoats"]
    },
    "Sportswear": {
        "subcategories": ["Running Apparel", "Yoga Pants", "Sports Bras", "Cycling Apparel", "Swimming Gear", "Training Shorts", "Compression Wear", "Tracksuits", "Sports Jackets", "Tennis Skirts"],
        "materials": ["Polyester", "Nylon", "Spandex", "Cotton", "Mesh"],
        "keywords": ["Performance", "Comfortable", "Breathable", "Flexible", "Active", "Training", "Fitness", "Stylish", "Gym Ready", "Sportswear"],
        "hashtags": ["#sportswear", "#fitness", "#active", "#athletic", "#sports", "#running", "#gymwear", "#yogapants"]
    },
    "Luxury Fashion": {
        "subcategories": ["Luxury Handbags", "Designer Shoes", "High-End Apparel", "Luxury Watches", "Jewelry", "Designer Sunglasses", "Tailored Suits", "Silk Scarves"],
        "materials": ["Leather", "Silk", "Gold", "Silver", "Stainless Steel", "Crystal", "Velvet"],
        "keywords": ["Luxury", "Premium", "Designer", "Exclusive", "Tailored", "Elegant", "Chic", "Classy", "Sophisticated", "Couture"],
        "hashtags": ["#luxury", "#highfashion", "#designer", "#elegant", "#chic", "#couture", "#exclusive", "#tailored", "#premium"]
    },
    "Sustainable Fashion": {
        "subcategories": ["Eco-friendly Footwear", "Recycled Clothing", "Organic Cotton Apparel", "Vegan Leather Bags", "Sustainable Activewear", "Upcycled Fashion", "Eco-Friendly Accessories", "Ethically Made Clothing"],
        "materials": ["Organic Cotton", "Recycled Polyester", "Vegan Leather", "Bamboo", "Hemp", "Upcycled Materials", "Natural Dyes"],
        "keywords": ["Eco-friendly", "Sustainable", "Ethical", "Organic", "Fair Trade", "Vegan", "Green", "Recycled", "Upcycled", "Circular Fashion"],
        "hashtags": ["#sustainablefashion", "#ecofriendly", "#organic", "#ethicalfashion", "#upcycled", "#greenfashion", "#fairtrade", "#circularfashion"]
    }
}

fashion_trend_data_new = []

for i in range(total_records):
    product_id = f"PROD{random.randint(1000, 9999)}"
    review_id = f"REV{random.randint(10000, 99999)}"
    social_post_id = f"POST{random.randint(10000, 99999)}"
    customer_id = f"CUST{random.randint(1000, 9999)}"
    competitor_id = f"COMP{random.randint(1000, 9999)}"
    campaign_id = f"CAM{random.randint(1000, 9999)}"

    # Product and Category Information
    category = random.choice(list(category_subcategories_materials_keywords_hashtags.keys()))  # Select random category
    subcategory = random.choice(category_subcategories_materials_keywords_hashtags[category]["subcategories"])  # Select subcategory
    material = random.choice(category_subcategories_materials_keywords_hashtags[category]["materials"])  # Select material based on category
    keywords = random.choices(category_subcategories_materials_keywords_hashtags[category]["keywords"], k=3)  # Select random keywords
    hashtags = random.choices(category_subcategories_materials_keywords_hashtags[category]["hashtags"], k=3)  # Select random hashtags
    product_name = f"Product {random.randint(1, 1000)}"
    brand = random.choice(["Skechers", "Converse", "Reebok", "Clarks", "Birkenstock", "Apple", "Sony", "Nike", "Samsung", "Adidas"])
    launch_date = random_date(start_date, end_date).strftime('%Y-%m-%d')
    seasonality = random.choice(["Summer", "Winter", "All Season"])
    eco_friendly_flag = random.choice([True, False])
    sustainability_score = random_float(3.0, 5.0) if eco_friendly_flag else random_float(1.5, 3.0)

    # Sentiment and Review Data
    review_date = random_date(start_date, end_date).strftime('%Y-%m-%d')
    rating = random_float(1.0, 5.0)

    # Adjust sentiment based on rating
    if rating < 2.5:
        sentiment = 'negative'
        sentiment_score = random_float(-1.0, -0.5)
        emotion_category = "Disappointed"
    elif rating > 3.5:
        sentiment = 'positive'
        sentiment_score = random_float(0.5, 1.0)
        emotion_category = "Happy"
    else:
        sentiment = 'neutral'
        sentiment_score = random_float(-0.5, 0.5)
        emotion_category = "Neutral"

    review_text = f"Review text for {product_name}, sentiment is {sentiment}."
    helpful_votes = random.randint(0, 50)
    photo_uploaded = random.choice([True, False])

    # Social Media and Influencer Data
    platform = random.choice(["Instagram", "Twitter", "TikTok", "Facebook"])
    engagement_count = random.randint(0, 10000) if sentiment == 'positive' else random.randint(0, 2000)
    post_date = random_date(start_date, end_date).strftime('%Y-%m-%d')
    post_content = f"Post about {product_name}, trend insights shared."
    influencer_id = f"INFL{random.randint(1000, 9999)}"
    influencer_score = random_float(3.0, 5.0)
    trend_score = random_float(0.0, 1.0)

    # Market and Competitor Data
    competitor_product_name = f"Competitor Product {random.randint(1, 1000)}"
    competitor_sentiment_score = random_float(-1.0, 1.0)
    market_popularity_index = random_float(0.0, 1.0)
    price_trend_comparison = random_float(-10.0, 10.0)

    # Customer Behavior and Demographics
    age_group = random.choice(["18-25", "26-35", "36-45", "46+"])
    gender = random.choice(["Male", "Female", "Other"])
    location = random.choice(["Jakarta", "Surabaya", "Bandung", "Medan", "Bali"])
    purchase_frequency = random.randint(1, 10)
    average_order_value = random_float(100000, 3000000) if eco_friendly_flag else random_float(50000, 1500000)

    # Sustainability Metrics
    eco_friendly_keyword_count = random.randint(0, 10)
    sustainability_sentiment_score = random_float(-1.0, 1.0)
    recycled_material_percentage = random_float(0.0, 100.0)
    carbon_footprint_impact = random_float(0.0, 10.0)

    # = = = Khusus Overall_Trend_Score = = =

    normalized_sentiment = (sentiment_score + 1) / 2

    # Purchase Frequency: 1–10 → 0 to 1
    normalized_purchase = (purchase_frequency - 1) / 9

    # Order Value: 50k–3jt → scale to 0–1
    normalized_order_value = (average_order_value - 50000) / (3000000 - 50000)

    # Engagement Count: 0–10,000 → 0 to 1
    normalized_engagement = engagement_count / 10000

    # Influencer Score: 3.0–5.0 → 0 to 1
    normalized_influencer = (influencer_score - 3.0) / 2.0

    # Trend Score: already 0–1
    normalized_trend = trend_score


    # Advanced Metrics and Trends
    overall_trend_score = (
    0.3 * normalized_sentiment +
    0.25 * normalized_purchase +
    0.15 * normalized_order_value +
    0.1 * normalized_engagement +
    0.1 * normalized_influencer +
    0.1 * normalized_trend
    )
    trend_longevity = random.randint(1, 365)
    emerging_trend_flag = random.choice([True, False])

    fashion_trend_data_new.append([
        product_id, product_name, category, subcategory, brand, launch_date, seasonality, material,
        eco_friendly_flag, sustainability_score, review_id, review_date, review_text, sentiment_score,
        emotion_category, keywords, rating, helpful_votes, photo_uploaded, social_post_id, platform,
        post_date, post_content, influencer_id, influencer_score, engagement_count, hashtags, trend_score,
        competitor_id, competitor_product_name, competitor_sentiment_score, market_popularity_index,
        price_trend_comparison, customer_id, age_group, gender, location, purchase_frequency,
        average_order_value, eco_friendly_keyword_count, sustainability_sentiment_score,
        recycled_material_percentage, carbon_footprint_impact, overall_trend_score, trend_longevity,
        emerging_trend_flag
    ])

# Define Columns for DataFrame
fashion_trend_columns = [
    "Product_ID", "Product_Name", "Category", "Subcategory", "Brand", "Launch_Date", "Seasonality",
    "Material", "Eco_Friendly_Flag", "Sustainability_Score", "Review_ID", "Review_Date", "Review_Text",
    "Sentiment_Score", "Emotion_Category", "Keyword_Tags", "Rating", "Helpful_Votes", "Photo_Uploaded",
    "Social_Post_ID", "Platform", "Post_Date", "Post_Content", "Influencer_ID", "Influencer_Score",
    "Engagement_Count", "Hashtags", "Trend_Score", "Competitor_ID", "Competitor_Product_Name",
    "Competitor_Sentiment_Score", "Market_Popularity_Index", "Price_Trend_Comparison", "Customer_ID",
    "Age_Group", "Gender", "Location", "Purchase_Frequency", "Average_Order_Value",
    "Eco_Friendly_Keyword_Count", "Sustainability_Sentiment_Score", "Recycled_Material_Percentage",
    "Carbon_Footprint_Impact", "Overall_Trend_Score", "Trend_Longevity", "Emerging_Trend_Flag"
]

# Create DataFrame
fashion_trend_df_new = pd.DataFrame(fashion_trend_data_new, columns=fashion_trend_columns)
fashion_trend_df_new.head()

```
