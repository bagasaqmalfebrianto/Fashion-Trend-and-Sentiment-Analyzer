# Grafik Trend Score

```python
df['Launch_Date'] = pd.to_datetime(df['Launch_Date'])
df['Tahun'] = df['Launch_Date'].dt.to_period('M').astype(str)
trend_score = df.groupby('Tahun')['Trend_Score'].mean().reset_index()
trend_score
```

```python
fig, ax = plt.subplots(figsize=(10, 3))  # Ukuran sesuai height=280

ax.plot(trend_score['Tahun'], trend_score['Trend_Score'], marker='o', linestyle='-')
ax.set_xlabel('Bulan (YYYY-MM)')
ax.set_ylabel('Rata-rata Trend Score')
# ax.set_title('📈 Perkembangan Trend Score per Bulan')  # Jika perlu

# Format tampilan x-tick
plt.xticks(rotation=45)
plt.tight_layout()

# Tampilkan di Streamlit
plt.show()
```

# Top 5 Brand

```python
brand_top5 = df.groupby(by='Brand').agg({
    'Overall_Trend_Score': 'mean'
}).sort_values(by='Overall_Trend_Score', ascending=False).head(5).reset_index()


brand_top5
```

```python
def bar_chart(data, col, color, title, x_label):
    plt.figure(figsize=(8, 5))
    plt.bar(data, col, color=color)
    plt.yscale('log')
    plt.title(title)
    plt.xlabel(x_label)
    plt.gca().axes.get_yaxis().set_visible(False)
    plt.ylabel('')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python

bar_chart(brand_top5['Brand'], brand_top5['Overall_Trend_Score'], '#0328fc','Top 5 Brand By Overall Trend Score', 'Brand' )
```

# WordCloud

```python
from wordcloud import WordCloud

hastags_series = df['Hashtags']

hashtags_text = " ".join([" ".join(h) if isinstance(h, list) else h for h in hastags_series])
```

```python
wordcloud = WordCloud(
    width=800, 
    height=400, 
    background_color='white', 
    colormap='tab10', 
    max_words=100
).generate(hashtags_text)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("", fontsize=16)
plt.show()
```
