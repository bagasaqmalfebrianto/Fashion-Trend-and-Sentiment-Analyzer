# After Perbaikan

## Hubungan Antara Sentimen dan Rating

### 1. Sentimen Positif
Jika sentimen positif, pastikan rating lebih tinggi, misalnya antara **4.0 hingga 5.0**.

### 2. Sentimen Negatif
Jika sentimen negatif, pastikan rating lebih rendah, misalnya antara **1.0 hingga 2.0**.

### 3. Sentimen Netral
Untuk sentimen netral, pastikan rating berada di sekitar nilai **3.0**.

---

## Penyesuaian Sentimen Berdasarkan Ulasan dan Rating

### 1. Rating di bawah 2.5
Jika rating di bawah **2.5**, tentukan sentimen sebagai **negatif**.

### 2. Rating lebih dari 3.5
Jika rating lebih dari **3.5**, tentukan sentimen sebagai **positif**.

### 3. Rating sekitar 3.0
Untuk rating sekitar **3.0**, gunakan sentimen **netral**.

---

## Sentimen Produk di Platform Sosial Media

Jika ada banyak postingan atau engagement dengan **sentimen positif**, dapatkan korelasi antara sentimen produk di review dan sentimen platform.

---

##Pencocokan Emosi dan Sentimen

- **Sentimen Positif** harus terkait dengan emosi seperti **"Happy"** atau **"Excited"**.
- **Sentimen Negatif** harus terkait dengan emosi seperti **"Disappointed"**.
- **Sentimen Netral** dapat dikaitkan dengan emosi **"Neutral"**.

# Pandas Agent with Prompt

```python
from langchain.prompts import PromptTemplate

prompt_template = PromptTemplate.from_template("""
Kamu adalah asisten analis data fashion. Tugasmu adalah menerjemahkan pertanyaan pengguna menjadi instruksi eksplisit analisis data Pandas.

Contoh:
- "Apa produk paling booming tahun 2023?" → "Filter Emerging_trend_flag=True, lalu urutkan Overall_Trend_Score tertinggi di tahun 2023"
- "Brand eco-friendly paling populer?" → "Filter Eco_Friendly_Flag=True, lalu urutkan berdasarkan Engagement_Count"
- "Tren kategori paling positif tahun 2022?" → "Group by Category pada tahun 2022, hitung rata-rata Sentiment_Score dan urutkan"

Gunakan hanya kolom berikut:
Product_Name, Category, Subcategory, Brand, Launch_Date, Seasonality, Material, Eco_Friendly_Flag,
Sustainability_Score, Review_Date, Review_Text, Sentiment_Score, Emotion_Category, Keyword_Tags,
Rating, Helpful_Votes, Engagement_Count, Hashtags, Trend_Score, Market_Popularity_Index,
Location, Purchase_Frequency, Average_Order_Value, Overall_Trend_Score, Trend_Longevity, Emerging_trend_flag

Sekarang, terjemahkan pertanyaan berikut:
Pertanyaan: {user_question}
Instruksi:
""")
```

```python
def hybrid_rag_pandas(user_question):
    # Step 1: Interpretasi oleh RAG
    interpret_prompt = prompt_template.format(user_question=user_question)
    interpreted_instruction = llm.predict(interpret_prompt)

    print("Interpretasi RAG:", interpreted_instruction)

    result = pandas_agent.invoke({"input": interpreted_instruction})

    final_output = result["output"]

    return final_output
```

```python
# 3. Contoh Pertanyaan
question = "Apa kategori produk yang paling trend di tahun 2023?"
answer = hybrid_rag_pandas(question)

print("\nJawaban Akhir:")
print(answer)
```
